# Plot tutorial

## Basic Plotting: Line Plots

```julia
using Plots
x = 1:10; y = rand(10); # These are the plotting data
plot(x, y)
```

```julia
x = 1:10; y = rand(10, 2) # 2 columns means two lines
plot(x, y)
```

```julia
z = rand(10)
plot!(x, z)
```

```julia
x = 1:10; y = rand(10, 2) # 2 columns means two lines
p = plot(x, y)
z = rand(10)
plot!(p, x, z)
```

## Plot Attributes

```julia
x = 1:10; y = rand(10, 2) # 2 columns means two lines
plot(x, y, title = "Two Lines", label = ["Line 1" "Line 2"], lw = 3)
```

```julia
xlabel!("My x label")
```

```julia
ylabel!("My y label")
```
